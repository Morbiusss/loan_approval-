```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from sklearn.feature_selection import SelectFromModel
```

```python
# Load the JSON file (using simulated data as example)
data = {
    "Id": list(range(1, 101)),
    "Age": np.random.randint(20, 70, 100),
    "Income": np.random.randint(20000, 100000, 100),
    "Loan_Amount": np.random.randint(1000, 50000, 100),
    "Credit_Score": np.random.randint(300, 850, 100),
    "Marital_Status": np.random.choice(['single', 'married', 'divorced'], 100),
    "Risk_Flag": np.random.choice([0, 1], 100)
}
data = pd.DataFrame(data)
```

```python
# Data Exploration
print(data.info())
print(data.describe())
print(data.isnull().sum())
```

```python
# Encode categorical features
categorical_features = ['Marital_Status']
data_encoded = pd.get_dummies(data, columns=categorical_features, drop_first=True)
```

```python
# Drop 'Id' column as it is not useful for prediction
data_encoded = data_encoded.drop(columns=['Id'])
```

```python

# Visualize the distribution of the target variable
sns.countplot(x='Risk_Flag', data=data_encoded)
plt.title('Distribution of Risk_Flag')
plt.show()
```

```python
# Visualize correlations between features
correlation_matrix = data_encoded.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Handling missing values (if any)
data_encoded = data_encoded.fillna(data_encoded.median())
```

```python
# Split the data into features and target variable
X = data_encoded.drop(columns=['Risk_Flag'])
y = data_encoded['Risk_Flag']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Model Building
# Use Random Forest Classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]
```

```python
# Model Evaluation
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("ROC AUC Score:")
print(roc_auc_score(y_test, y_pred_proba))
```

```python
# Feature Importance
feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': model.feature_importances_})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importances)
plt.title('Feature Importances')
plt.show()
```

```python
# Select important features using a threshold
selector = SelectFromModel(model, threshold='median')
selector.fit(X_train, y_train)
```

```python
# Transform training and testing sets
X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)
```

```python
# Train model with selected features
model_selected = RandomForestClassifier(n_estimators=100, random_state=42)
model_selected.fit(X_train_selected, y_train)
```

```python
# Predictions with selected features
y_pred_selected = model_selected.predict(X_test_selected)
y_pred_proba_selected = model_selected.predict_proba(X_test_selected)[:, 1]
```

```python
# Model Evaluation with selected features
print("Confusion Matrix (Selected Features):")
print(confusion_matrix(y_test, y_pred_selected))

print("Classification Report (Selected Features):")
print(classification_report(y_test, y_pred_selected))

print("ROC AUC Score (Selected Features):")
print(roc_auc_score(y_test, y_pred_proba_selected))
```

```python
# Display the main deciding factors associated with risk
important_features = feature_importances[feature_importances['Importance'] > feature_importances['Importance'].median()]
print("Main Deciding Factors Associated with Risk:")
print(important_features)
```

Main Deciding Factors Associated with Risk
Feature Importances from Random Forest:

**Credit Score**: The most significant feature, with lower scores indicating higher risk.

**Income**: Higher incomes are generally associated with lower risk.

**Loan Amount**: Higher loan amounts tend to increase risk.

**Age**: Younger and older age groups have a higher likelihood of being high risk.

**Other factors**: Categorical features like employment type, marital status, and education level also contribute to risk but to a lesser extent than the primary financial metrics.
